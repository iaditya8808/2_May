```python
df = spark.read.csv("Files/sales_data.csv", header=True, inferSchema=True)

customer_df = spark.read.csv("Files/customer_data.csv", header=True, inferSchema=True)

product_df = spark.read.csv("Files/product_data.csv", header=True, inferSchema=True)

final_df = df.join(customer_df, "customer_id") \
             .join(product_df, "product_id")

final_df.show()
```

```python
final_df.createOrReplaceTempView("sales")

spark.sql("""
SELECT product_id, SUM(price*quantity) as revenue
FROM sales
GROUP BY product_id
ORDER BY revenue DESC
LIMIT 5
""").show()
```

```python
from pyspark.sql.functions import month

final_df.groupBy(month("timestamp").alias("month")) \
        .sum("price").show()
```
